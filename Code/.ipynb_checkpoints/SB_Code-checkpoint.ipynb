```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

Data_csv_path = "../Data/Employee_Table.csv"

first_df =  pd.read_csv(Data_csv_path)
first_df.columns
```

```python
ET = pd.DataFrame(first_df[["AIdent", "Branch", 'City', 'How_Heard']])
ET.head(10)
```

```python
Data_csv_path2 = "../Data/Assignment_Table.csv"

A_df =  pd.read_csv(Data_csv_path2)
A_df.columns
```

```python
AS = pd.DataFrame(A_df[['AIdent', 'Activation', 'Completion', 'Pay_Rate', 'Bill_Rate']])
AS.head(10)
```

```python
Margin = []
Margin = AS["Bill_Rate"] - AS["Pay_Rate"]
Margin
AS["Margin"] = Margin

AS.min()
```

```python
Data_csv_path3 = "../Data/Hours_Worked.csv"

HW_df =  pd.read_csv(Data_csv_path3)
HW_df.columns
```

```python
HW = pd.DataFrame(HW_df[['AIdent', 'Hours_Worked']])

HW.head(10)
```

```python
unique_ET = ET["AIdent"].nunique()
unique_AS = AS["AIdent"].nunique()
unique_HW = HW["AIdent"].nunique()

count_ET = ET["AIdent"].count()
count_AS = AS["AIdent"].count()
count_HW = HW["AIdent"].count()

print("Number of unique AIdents")
print(unique_ET)
print(unique_AS)
print(unique_HW)
print("-------------------")
print("Count of AIdents")
print(count_ET)
print(count_AS)
print(count_HW)
```
